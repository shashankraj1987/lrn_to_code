Suppose that you are given two strings s and t. String t is generated by random shuffling string s and then adding one more letter at a random position. Return the letter that was added to t. 

Example 1:
Input: s = "abcd", t = "ceadb"
output = "e"

Exmaple 2:
Input: s = "abbdd", t = "dabadb"
output = "a"

Example 3: 
Input: s = "aaaa", t = "aabaa"
output = "b"

Example 4:
Input: s = "ae", t = "aae"
output = "a"


```python
def findTheDifference(s: str,t:str) -> str:
    dict_s = {}
    dict_t = {}

    for i in s:
        if i in dict_s:
            dict_s[i] += 1
        else:
            dict_s[i] = 1

    for i in t:
        if i in dict_t:
            dict_t[i] += 1
        else:
            dict_t[i] = 1

    for key in dict_t.keys():
        if key in dict_s.keys():
            result = '-'
        else:
            result = key
            break

    if result == "-":
        for key in dict_t.keys():
            if dict_t[key] == dict_s[key]:
                result = '-'
            else:
                result = key
                return result
    else:
        return result

print(findTheDifference("ae","aae"))
```
