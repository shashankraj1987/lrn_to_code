```python
ops = ["5","-2","4","C","D","9","+","+"]

import re
final_arr = []
final_val = 0
import logging

logging.basicConfig(level=logging.INFO)

for x in ops:
    if re.findall("[0-9]",x):
        final_arr.append(int(x))
        logging.info(f'Found Number, adding {x}')
        logging.info(f'Final Array is {final_arr}')
    elif x == "C":
        logging.info(f'Found C, Popping')
        logging.info(f'Final Array is {final_arr}')
        final_arr.pop()
    elif x == "D":
        final_arr.append(int(final_arr[-1]*2))
        logging.info(f'Found D, Multiplying with *2 to {x}')
        logging.info(f'Final Array is {final_arr}')
    elif x == "+":
        final_arr.append(int(final_arr[-1])+int(final_arr[-2]))
        logging.info(f'Found +, adding {x} to previous val')
        logging.info(f'Final Array is {final_arr}')

print("*"*20)
for x in final_arr:
    final_val+=x

print(final_val)
```

```python
def process_arr(s):
    s_brk = 0
    c_brk = 0
    sqr_brk = 0
    invalid = 0

    if len(s) <= 104 and len(s)>1:
        pass 
    else:
        raise ValueError

    for x in s:
        if x == "(":
            s_brk += 1
        elif x == ")":
            s_brk += 1
        elif x == "[":
            sqr_brk += 1
        elif x == "]":
            sqr_brk += 1
        elif x == "{":
            c_brk += 1
        elif x == "}":
            c_brk += 1
        else:
            invalid = 1

    if s_brk%2 == 0 and c_brk%2 == 0 and sqr_brk%2 == 0 and invalid == 0:
        return "Valid"
    else:
        return "Invalid"

def main():
    arr = input()
    print(process_arr(arr))

if __name__ == "__main__":
    main()
```
